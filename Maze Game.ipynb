```python
import turtle

GRID_SIZE = 24
walls = []

# register shapes 
turtle.register_shape('/Users/ameliacalantzis/coins/pankeeki.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/orange.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/pizza.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/pink.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/star.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/camel.gif')
turtle.register_shape('/Users/ameliacalantzis/coins/waterbottle.gif')

screen = turtle.Screen()
screen.bgcolor('light blue')
screen.bgpic('/Users/ameliacalantzis/coins/moon.png')
player = turtle.Turtle()
player.shape('/Users/ameliacalantzis/coins/pink.gif')
player.color('white')
player.penup()
player.goto(-288, 120)

# create gold treasure
treasure = turtle.Turtle()
treasure.color('gold')
treasure.shape('/Users/ameliacalantzis/coins/orange.gif')
treasure.penup()
treasure.goto(-200, 108)

# create green treasure
greentreasure = turtle.Turtle()
greentreasure.color('green')
greentreasure.shape('/Users/ameliacalantzis/coins/pankeeki.gif')
greentreasure.penup()
greentreasure.goto(160, 110)

# create blue treasure
bluetreasure = turtle.Turtle()
bluetreasure.color('blue')
bluetreasure.shape('/Users/ameliacalantzis/coins/pizza.gif')
bluetreasure.penup()
bluetreasure.goto(60,-230)

def treasure_collision_check():
    if player.xcor() > treasure.xcor() - GRID_SIZE and player.xcor() < treasure.xcor() + GRID_SIZE and player.ycor() > treasure.ycor() - GRID_SIZE and player.ycor() < treasure.ycor() + GRID_SIZE:
        treasure.hideturtle()

def greentreasure_collision_check():
    if player.xcor() > greentreasure.xcor() - GRID_SIZE and player.xcor() < greentreasure.xcor() + GRID_SIZE and player.ycor() > greentreasure.ycor() - GRID_SIZE and player.ycor() < greentreasure.ycor() + GRID_SIZE:
        greentreasure.hideturtle()
        
def bluetreasure_collision_check():
    if player.xcor() > bluetreasure.xcor() - GRID_SIZE and player.xcor() < bluetreasure.xcor() + GRID_SIZE and player.ycor() > bluetreasure.ycor() - GRID_SIZE and player.ycor() < bluetreasure.ycor() + GRID_SIZE:
        bluetreasure.hideturtle()
        
def end_maze_check():
    if player.xcor() >= 288 and player.xcor() <= 312 and player.ycor() >= -48 and player.ycor() <= -24:
        turtle.resetscreen()
        screen.bgpic('/Users/ameliacalantzis/coins/properegypt.png')
        walls.clear()
        turtle.onkey(None, 'Up')
        turtle.onkey(None, 'Down')
        turtle.onkey(None, 'Right')
        turtle.onkey(None, 'Left')
        setup_maze(level_2)
        
        player.shape('/Users/ameliacalantzis/coins/camel.gif')
        player.color('white')
        player.penup()
        player.goto(-288, 0)

        # create gold treasure
        treasure.color('gold')
        treasure.shape('/Users/ameliacalantzis/coins/waterbottle.gif')
        treasure.penup()
        treasure.goto(-190, 108)

        # create green treasure
        greentreasure.color('green')
        greentreasure.shape('/Users/ameliacalantzis/coins/waterbottle.gif')
        greentreasure.penup()
        greentreasure.goto(180, -40)

        # create blue treasure
        bluetreasure.color('blue')
        bluetreasure.shape('/Users/ameliacalantzis/coins/waterbottle.gif')
        bluetreasure.penup()
        bluetreasure.goto(48,-230)
        
        turtle.listen()
        turtle.onkey(up_L2, 'Up')
        turtle.onkey(down_L2, 'Down')
        turtle.onkey(right_L2, 'Right')
        turtle.onkey(left_L2, 'Left')

def end_level_2():
    if player.xcor() >= 288 and player.xcor() <= 312 and player.ycor() >= -288 and player.ycor() <= --264:
        turtle.clearscreen()
        turtle.write("Great job! You finished the game!", align="center", font=("Verdana", 40, "normal"))
        
def move(shift_x, shift_y):
    """ Move Turtle
    :param shift_x: Shift in X coordinate
    :param shift_y: Shift in Y coordinate """
    x = player.xcor() + shift_x
    y = player.ycor() + shift_y
    if can_turtle_go(x, y):
        player.setx(x)
        player.sety(y)
    treasure_collision_check()
    greentreasure_collision_check()
    bluetreasure_collision_check()
    end_maze_check()

def can_turtle_go(x, y):
    """ Check if Turtle can go
    :param x: next X coord for Player
    :param y: next Y coord for Player
    :return: True - Turtle can go
             False - Turtle can't go """
    for wall in walls:
        if x == wall[0] and y == wall[1]:
            return False  # Turtle can't go
    return True  # Turtle can go

def up():
    move(0, GRID_SIZE)

def down():
    move(0, -GRID_SIZE)

def right():
    move(GRID_SIZE, 0)

def left():
    move(-GRID_SIZE, 0)
    
def setup_maze(level):
    wall = turtle.Turtle()
    wall.shape('/Users/ameliacalantzis/coins/star.gif')
    wall.color('white')
    wall.penup()
    wall.speed(0)

    for y in range(len(level)):
        for x in range(len(level[y])):
            character = level[y][x]
            screen_x = -288 + (x * GRID_SIZE)
            screen_y = 288 - (y * GRID_SIZE)

            if character == 'X':
                walls.append((screen_x, screen_y))
                wall.goto(screen_x, screen_y)
                wall.stamp()
                
setup_maze(level_1)
turtle.listen()
turtle.onkey(up, 'Up')
turtle.onkey(down, 'Down')
turtle.onkey(right, 'Right')
turtle.onkey(left, 'Left')
player.penup()

def move_L2(shift_x, shift_y):
    """ Move Turtle
    :param shift_x: Shift in X coordinate
    :param shift_y: Shift in Y coordinate """
    x = player.xcor() + shift_x
    y = player.ycor() + shift_y
    if can_turtle_go(x, y):
        player.setx(x)
        player.sety(y)
    treasure_collision_check()
    greentreasure_collision_check()
    bluetreasure_collision_check()
    end_level_2()

def up_L2():
    move_L2(0, GRID_SIZE)

def down_L2():
    move_L2(0, -GRID_SIZE)

def right_L2():
    move_L2(GRID_SIZE, 0)

def left_L2():
    move_L2(-GRID_SIZE, 0)

turtle.done()
```
